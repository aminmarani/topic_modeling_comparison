# comparing NPMI of top-N labeler and factors

1. loading top-N labeler factor values

```python
data = pd.read_csv('./topic_labeling/data_with_2factors',index_col=0)
data
```

```python
#loading data that includes topicID, top_n words , and top documents
data = pd.read_csv('./topic_labeling/topics.csv',index_col=0,sep='\t;')
data = data.loc[data['Input.topicid'].str.len()<5,]
data = data.rename(columns={'Input.topicid':'topicID'})
data.topicID = pd.to_numeric(data.topicID)

data = pd.merge(pd.read_csv('./topic_labeling/data_with_3factors.csv',index_col=0,sep='\t'),
             data,on=['corpus','topicID'])
data = data.drop_duplicates()
data
```

2. only keeping topN technique

```python
data = data[data.technique == 'TopN']
data_avg = data.groupby(by=['corpus','topicID','technique','Input.topn', 'Input.doc1', 'Input.doc2', 'Input.doc3', 'Input.doc4',
       'Input.doc5']).mean()
data_avg = data_avg.reset_index()
data_avg
```

3. computing NPMI for top-n words

```python
from lda_mallet import lda_score
import pickle

score = lda_score(num_topics=10,alpha=10,optimize_interval=10,iterations=1000,wiki_path='./data/wiki_sampled_10p.txt',
                  db_path = './db/wiki_full/',vocab_dict_path = './data/corpus.vocab', 
                  wiki_vocab_dict_path='./data/ref_corpus.vocab',npmi_skip_threshold=0.05)
score.all_top_terms = [[i.strip() for i in s.strip().split(',') if i.strip() not in ['corp']] for s in data_avg['Input.topn']]

with open('./data/corpus.vocab','rb') as f:
    score.wiki_vocab_dict = pickle.load(f)


res = score.score([])
res
# # res = score.fit(X)
```
