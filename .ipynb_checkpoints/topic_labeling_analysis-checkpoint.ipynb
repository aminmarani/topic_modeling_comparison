# comparing NPMI of top-N labeler and factors

1. loading top-N labeler factor values

```python
data = pd.read_csv('./topic_labeling/data_with_2factors',index_col=0)
data
```

```python
#loading data that includes topicID, top_n words , and top documents
data = pd.read_csv('./topic_labeling/topics.csv',index_col=0,sep='\t;')
data = data.loc[data['Input.topicid'].str.len()<5,]
data = data.rename(columns={'Input.topicid':'topicID'})
data.topicID = pd.to_numeric(data.topicID)

data = pd.merge(pd.read_csv('./topic_labeling/data_with_3factors.csv',index_col=0,sep='\t'),
             data,on=['corpus','topicID'])
data = data.drop_duplicates()
data
```

2. only keeping topN technique

```python
data = data[data.technique == 'TopN']
data_avg = data.groupby(by=['corpus','topicID','technique','Input.topn', 'Input.doc1', 'Input.doc2', 'Input.doc3', 'Input.doc4',
       'Input.doc5']).mean()
data_avg = data_avg.reset_index()
data_avg
```

3. computing NPMI for top-n words

```python
from lda_mallet import lda_score, loading_wiki_docs
import pickle
from gensim.models.coherencemodel import CoherenceModel


# score = lda_score(num_topics=10,alpha=10,optimize_interval=10,iterations=1000,wiki_path='../wiki_corpus/wiki_full',
#                   db_path = './data/wiki_full/',vocab_dict_path = './data/corpus.vocab', 
#                   wiki_vocab_dict_path='./data/ref_corpus.vocab',npmi_skip_threshold=0.05)

# with open('./data/wiki_full_vocab.obj','rb') as f:
#     score.wiki_vocab_dict = pickle.load(f)
    
all_top_terms = [[i.strip() for i in s.strip().split(',') if i.strip() in score.wiki_vocab_dict.token2id] for s in data_avg['Input.topn']]

##Read
with open('/home/amh418/erb416_proj/amh418/wiki_corpus/wiki_full_pre_processed_docs.list','rb') as objfile:
    pre_processed_wiki = pickle.load(objfile)
    
with open('/home/amh418/erb416_proj/amh418/wiki_corpus/wiki_full_vocab.obj','rb') as objfile:
    wiki_vocab_dict = pickle.load(objfile)

# cscore_1 = CoherenceModel(topics=all_top_terms,dictionary=wiki_vocab_dict,texts=pre_processed_wiki,coherence='c_npmi',processes=6,topn=3).get_coherence_per_topic()
# cscore_2 = CoherenceModel(topics=all_top_terms,dictionary=wiki_vocab_dict,texts=pre_processed_wiki,coherence='c_npmi',processes=6,topn=6).get_coherence_per_topic()
# CoherenceModel(topics=all_top_terms,dictionary=ref_dict,texts=texts,coherence='c_npmi',processes=3).get_coherence_per_topic()

# res = score.score([])
# res
# # res = score.fit(X)
```

```python
from gensim.models.coherencemodel import CoherenceModel

cscore_1 = CoherenceModel(topics=all_top_terms,dictionary=wiki_vocab_dict,texts=pre_processed_wiki,coherence='c_npmi',processes=6,topn=3).get_coherence_per_topic()
cscore_2 = CoherenceModel(topics=all_top_terms,dictionary=wiki_vocab_dict,texts=pre_processed_wiki,coherence='c_npmi',processes=6,topn=6).get_coherence_per_topic()
```

```python
data_avg['NPMI_3'] = cscore_1
data_avg['NPMI_6'] = cscore_2
data_avg['NPMI'] = [(cscore_1[i]+cscore_2[i])/2 for i in range(len(data_avg))]
data_avg.to_csv('./topic_labeling//data_avg_with_NPMI.csv')
data_avg
```

3. Correlation of NPMI and factors

```python
from scipy.stats.stats import pearsonr

#dropping NA
df = data_avg[~pd.isna(data_avg.NPMI)]

print(pearsonr(df.suitable,df.NPMI))
print(pearsonr(df.suitable[df.corpus == 'ap'],df.NPMI[df.corpus == 'ap']))
print(pearsonr(df.preferable,df.NPMI))
print(pearsonr(df.objectionable,df.NPMI))
print(pearsonr(df.consensus,df.NPMI))
```
